#### Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all ?
**Ans:** A probability distribution is a mathematical function that describes the likelihood of different outcomes in a random experiment or event. It provides a set of probabilities for all possible outcomes, indicating how likely each outcome is to occur. While the values in a probability distribution are based on randomness, the distribution itself follows a defined pattern that can be predicted using probability theory and statistics.

#### Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?
**Ans:** : Yes, there is a distinction between true random numbers and pseudo-random numbers. True random numbers are generated from a truly random process, such as physical phenomena like radioactive decay or atmospheric noise, and they are unpredictable by nature. Pseudo-random numbers, on the other hand, are generated by algorithms and are deterministic, meaning they are based on an initial value called the seed. Given the same seed, a pseudo-random number generator will produce the same sequence of numbers. 

#### Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution ?
**Ans:** The two main factors that influence the behavior of a "normal" probability distribution are the mean (average) and the standard deviation. The mean represents the central tendency of the distribution, determining its peak or center, while the standard deviation measures the spread or dispersion of the values around the mean. Together, they define the shape and characteristics of the normal distribution.

#### Q5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows ?
**Ans:**  In the short term, the behavior of a probability distribution can be unpredictable due to random variations. However, as the number of trials or observations increases, the distribution tends to stabilize, and the probabilities of different outcomes become more consistent and closer to the expected theoretical probabilities. This is known as the law of large numbers, which states that with a large enough sample size, the observed frequencies of outcomes will converge to the true probabilities.

**Ans:** Python supports three logarithmic functions: math.log(x) computes the natural logarithm (base e) of x, math.log10(x) calculates the logarithm with base 10 of x, and math.log(x, base) can be used to compute the logarithm of x with a specified base.
